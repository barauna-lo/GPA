```python
import numpy as np
import plotly.graph_objects as go
from astroML.datasets import fetch_great_wall, fetch_nasa_atlas
from matplotlib import pyplot as plt
```

```python
x = fetch_nasa_atlas()
```

```python
def polar2cart(z, ra, dec):
    return z * np.sin(ra) * np.cos(dec),z * np.sin(ra) * np.sin(dec),z * np.cos(ra)
```

```python
import matplotlib.pyplot as plt

cx,cy,cz = polar2cart(x['Z'],x['RA'],x['DEC'])
n = np.random.choice(np.arange(len(x)),14000)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(cx[n], cy[n], cz[n])
plt.show()
```

```python
vol = np.load("nasaAtlasKDE.npy")
print(np.min(vol),np.max(vol))
```

```python
l = 32
X, Y, Z = np.mgrid[:l, :l, :l]
isoMin,isoMax = np.average(vol),np.max(vol)

title = "Atlas KDE 32x32x32"
fig = go.Figure(data=go.Volume(
    x=X.flatten(), y=Y.flatten(), z=Z.flatten(),
    value=vol.flatten(),
    isomin=isoMin,
    isomax=isoMax,
    opacity=0.1,
    surface_count=25,
    colorscale='Rainbow'
    ))
fig.update_layout(width=700, height=700,title = title)
fig.update_layout(scene = dict(
                    xaxis_title='RA',
                    yaxis_title='Z',
                    zaxis_title='DEC'),
                    margin=dict(r=20, b=10, l=10, t=10))

fig.show()
```
